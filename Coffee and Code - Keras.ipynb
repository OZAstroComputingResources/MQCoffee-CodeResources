### Keras and Tensorflow


Here we import Sequential from keras.models, which is a standard feedforward network.
Dense and Activation are layers of the network.
np_utils is useful later for converting numbers into one-hot vectors.


```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import np_utils
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
```

#### Basic Categorisation
We are setting up a dataset here, with two crescent shaped groups, categorised differently.

```python
X, y = sklearn.datasets.make_moons(200, noise=0.25, random_state=0)
yc = y
y = y.reshape(y.size,1).astype(float)

fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], s=40, c=yc, cmap = plt.cm.Spectral, zorder = 1)
plt.show()
```

We want to be able to predict which group a new data point will belong to. Neural networks can do this very easily. <br>

We start by using the previously imported Sequential class. <br>

Just by having gone through this many times before, I choose to have an input layer with two neurons (for the two coordinates), 50 neurons in the hidden layer (it probably need not be this high, my neural net can do it with 10), and one in the output (this will be a number between 1 and 0, where 1 is red and 0 is blue. <br>

Activation functions in both layers are sigmoids. This was chosen mostly because it is common, and I don't want to change the data beforehand (as sigmoid outputs numbers in the range 0 to 1). <br>

Mean squared error is the error function I defined earlier. There are better error functions for other problems, but this will suffice here. Adam refers to adaptive moment estimation, which is essentially an algorithm which speeds up the process of learning.

```python
model = Sequential()
model.add(Dense(input_dim = 2, units = 50, activation='sigmoid'))
#model.add(Activation('sigmoid'))
model.add(Dense(input_dim = 50, units = 1, activation='sigmoid'))
#model.add(Activation('sigmoid'))
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

Okay, after compiling the network, we want to fit the data. This is done easily by calling the fit() method of the network.

```python
hist = model.fit(X,y,epochs=2500,batch_size=32,verbose=1)
```

Here I have just retained the history object output by fitting the network. It just contains information on the loss and accuracy of the learning process.

```python
fig, ax = plt.subplots()
ax1 = ax.twinx()
ax.plot(hist.history['acc'], 'b')
ax1.plot(hist.history['loss'], 'r')
plt.show()
```

The red region in the plot below is the region the network believes objects should fall into the red category, and the same for the blue. We can see that with these very noisy moons, the network does a pretty good job, but cannot entirely separate them. If we'd had more input parameters, perhaps we could have certainly separated the groups.

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], s=40, c=yc, cmap = plt.cm.Spectral, zorder = 1)
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

XX, YY = np.mgrid[xmin:xmax:200j, ymin:ymax:200j]
X_in = np.c_[XX.ravel(), YY.ravel()]

Z = model.predict(X_in)
#Z = destandardise(Z,meany,stdy)
Z = Z.reshape(XX.shape)

Z = np.where(Z > 0.5, 1., 0.)

ax.pcolormesh(XX, YY, Z, cmap=plt.cm.RdBu, zorder = 0)
plt.show()
```

#### More difficult categoristion - MNIST Handwritten Digit Dataset

This is a standard example for people new to neural networks. <br>

The dataset consists of 60,000 28 x 28 pixel images of handwritten digits, with the correct labels, which are generally used for training a network. There are also 10,000 test examples, which are used to check the network is performing as expected. While not necessarily a useful real world case by itself, it can be used to demonstrate understanding of how neural networks function, as well as benchmarking speed tests of networks, testing accuracy, etc. <br>

Luckily, keras comes with a routine to download the data.

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

First 5 images in the dataset. Clearly a 7, 2, 1, 0 and 4. I'll use these numbers to quickly test the network we set up.

```python
f, axarr = plt.subplots(1,6)
for i in range(6):
    axarr[i].imshow(X_test[i],cmap=plt.cm.Greys)
    axarr[i].xaxis.set_major_locator(plt.NullLocator())
    axarr[i].yaxis.set_major_locator(plt.NullLocator())
plt.show()
```

I flatten the data out to be a list of 784 pixel values for each image. This is how I have generally seen it done, but keras also allows us to have 2D input layers. You can read about this on your own.

```python
X_train_flat = X_train.reshape(X_train.shape[0], 784)
X_test_flat = X_test.reshape(X_test.shape[0], 784)

Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
```

Similar to before. How many neurons should we have in the input layer? How many in the output layer?

```python
num_model = Sequential()
num_model.add(Dense(input_dim = 784, units = 30, activation='sigmoid'))
#model.add(Activation('sigmoid'))
num_model.add(Dense(input_dim = 30, units = 10, activation='softmax'))
#model.add(Activation('sigmoid'))
num_model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
print(np.argmax(num_model.predict(X_test_flat[0:5]),axis=1))
```

Clearly the model does a terrible job at predicting the values of the images.

No better than random chance at the moment (this is literally true, as the weights of the neurons are assigned randomly at first).

Perhaps training will give a better result. As the dataset contains 60,000 images, we can run through it 30 times to get a reasonable result.

```python
hist = num_model.fit(X_train_flat, Y_train, epochs=30, batch_size=32)
```

```python
fig, ax = plt.subplots()
ax1 = ax.twinx()
ax.plot(hist.history['acc'], 'b')
ax1.plot(hist.history['loss'], 'r')
plt.show()
```

```python
print(np.argmax(num_model.predict(X_test_flat[0:6]),axis=1))
```

Clearly the code is doing much better now. According to the accuracy plot, it is categorising about 92% of images correctly. Not bad for a few minutes of work!

```python
num_model.evaluate(X_test_flat, Y_test)
```
